```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle, json

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, f1_score, make_scorer
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from imblearn.over_sampling import ADASYN
```

```python
X_train = pd.read_csv('data/X_train.csv', index_col=0)
X_test = pd.read_csv('data/X_test.csv', index_col=0)
X_val = pd.read_csv('data/X_val.csv', index_col=0)

y_train = pd.read_csv('data/y_train.csv', index_col=0)
y_test = pd.read_csv('data/y_test.csv', index_col=0)
y_val = pd.read_csv('data/y_val.csv', index_col=0)

df_feat = pd.read_csv('data/fs_data.csv', index_col=0)
target = 'pov6'
col_ls = df_feat['features'][df_feat['voting'] > 2].tolist()
print('number of features used', len(col_ls))
X_train.shape, X_test.shape, X_val.shape, X_train.shape, y_test.shape, y_val.shape
```

## Resample Using ADASYN

```python
adasyn = ADASYN(sampling_strategy='auto')
X_train_resample, y_train_resampled = adasyn.fit_resample(X_train[col_ls], y_train[target])
```

```python
def grid_searcher( X_train, X_test, y_train, y_test, model, save = True):

    
    full_name_dict = {'lr'       : 'Logistic Regression',
                      'dt'       : 'Decision Tree',
                      'rf'       : 'Random Forest Classifier', 
                      'gb'       : 'Gradient Boosting_classifier',
                      'agb'      : 'Adaboosting Classifier',
                     }

    
    param_dict = {'lr' : {'lr__solver':['liblinear','lbfgs'],
                          'lr__C':[0.1,1,10],
                           'lr__max_iter':[100,1000,3000]},
                  
                  'dt' : {'dt__max_depth':[None, 1, 2,3],
                          'dt__min_samples_split': [2,4,10],
                          'dt__min_samples_leaf': [2,3]},
                
                  'rf' : {'rf__n_estimators': [100],
                          'rf__max_depth': [None, 1, 2,3],
                          'rf__min_samples_split': [2,5,10],
                          'rf__min_samples_leaf': [2,3]},
                
                  'gb' : {'gb__max_depth':[3,4,5],
                          'gb__n_estimators':[100, 300],
                          'gb__subsample': [0.8,1.0]},
                  
                  'agb': {'agb__learning_rate':[0.8,1.0],
                          'agb__n_estimators':[50, 100],
                          'agb__algorithm': ['SAMME', 'SAMME.R']},
                  
                 
                 }
                  
    model_dict = {
                  'lr' : LogisticRegression(solver='lbfgs'), 
                  'dt' : DecisionTreeClassifier(),
                  'rf' : RandomForestClassifier(),
                  'gb' : GradientBoostingClassifier(),
                  'agb': AdaBoostClassifier(),
                 }

    #choosing model for modelling
    model_dict_1 ={
                  'lr' : LogisticRegression,
                  'dt' : DecisionTreeClassifier,
                  'rf' : RandomForestClassifier,
                  'gb' : GradientBoostingClassifier,
                  'agb': AdaBoostClassifier,
                 
                  }
    print(f'''
    ###########################
    {full_name_dict[model]} model
    ###########################    
    ''')
    pipe = Pipeline([(model, model_dict[model])])
    pipe_params = param_dict[model]
    f1_scorer = make_scorer(f1_score, average='weighted')
    grid = GridSearchCV(pipe, param_grid=pipe_params, cv=StratifiedKFold(n_splits=10), n_jobs=-1,scoring =f1_scorer, verbose = 1)

    grid.fit(X_train, y_train.values.ravel())
    scores = pd.DataFrame(grid.cv_results_['params'])
    test_score_grid = grid.score(X_test,y_test)
    print('Test score from grid',test_score_grid)

    print('Best score on training data',round(grid.best_score_,4))    
    print('Best Params',grid.best_params_)   
    string_remove = model + '__'
    parameter = dict((k.replace(string_remove,''),v) for k,v in grid.best_params_.items())

    choosen_model = model_dict_1[model](**parameter).fit(X_train,y_train)   
    print('*******Test Data*******')
    y_pred = choosen_model.predict(X_test)

    test_score = f1_score(y_test, y_pred, average='weighted',labels=np.unique(y_pred))
    print('Test score for test set',test_score)
    cm = confusion_matrix(y_test, y_pred)
    cm_display = ConfusionMatrixDisplay(confusion_matrix= cm,display_labels=choosen_model.classes_)
    cm_display.plot()
    plt.show()
    print(classification_report(y_test, y_pred))
    
    file_path = f'./model/{model}.pkl'
    filesave = open(file_path, 'wb')
    pickle.dump(choosen_model, filesave)
    print('Model File Save', file_path)
```

```python
model_ls = ['lr','dt', 'rf', 'gb','agb']

for i in model_ls:
    grid_searcher(X_train_resample[col_ls], X_test[col_ls], y_train_resampled, y_test,i)
```

## Predict testdata.csv

```python
file = open('./model/rf.pkl', 'rb')
rf = pickle.load(file)

with open('./data/mapping_dict.json', 'r') as f:
    map_dict = json.load(f)

test = pd.read_csv('./data/testdata.csv', index_col=0)
```
